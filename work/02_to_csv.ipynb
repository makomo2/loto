```python
import csv
import itertools
import sys

from argparse import ArgumentParser

from lib import database as db
from lib import training as tr
```

```python
def get_option():
    a = ArgumentParser()    
    a.add_argument('-k', '--kind', choices = ['mini', 'loto6', 'loto7'], required = True)
    a = a.parse_args()
    return a
```

```python
a = get_option()
kind = a.kind
loto = db.session.query(db.Loto).filter(db.Loto.kind == kind).order_by(db.Loto.times).all()
```

```python
num_classes = tr.get_num_classes(kind)
x_test = []
y_test = []
for no, l in enumerate(loto):
    
    if no < tr.x_column: continue

    x = tr.get_x_data(loto, no)

    numbers = list(map(int, l.numbers.split(',')))
    bonus   = list(map(int, l.bonus.split(',')))
    for i in range(1, num_classes + 1):
        if i in (numbers + bonus):
            x_test.append([l.date.month, l.date.day] + x)
            y_test.append([i])
```

```python
with open('csv/' + kind + '_x.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerows(x_test)

with open('csv/' + kind + '_y.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerows(y_test)
```
