```python
import csv
import itertools
import sys

from lib import training as tr
from lib import database as db
```

```python
loto = db.session.query(db.Loto).order_by(db.Loto.times).all()
x_test = []
y_test = []
```

```python
for no, l in enumerate(loto):
    
    if no < tr.before_num: continue

    numbers = [l.one, l.two, l.three, l.four, l.five, l.six, l.bonus_one]
    before  = []
    for i in range(tr.before_num):
        b = loto[no - tr.before_num + i]
        before += [b.one, b.two, b.three, b.four, b.five, b.six]

    for i in range(1, tr.num_classes + 1):
        if i in numbers:
            x_test.append([l.date.month, l.date.day] + before)
            y_test.append([i])
```

```python
with open('csv/loto_x.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerows(x_test)

with open('csv/loto_y.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerows(y_test)
```
